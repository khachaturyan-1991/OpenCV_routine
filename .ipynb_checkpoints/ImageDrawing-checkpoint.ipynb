# Beggining

```python
def read_img(image, size=False, color = cv2.COLOR_RGB2BGR):
    img = cv2.imread(image)
    if color:
        try:
            img = cv2.cvtColor(img, color)
        except:
            raise Exception("Provide color in a format like cv2.COLOR_RGB2BGR")
    if size:
        try:
            s1,s2 = size
            img = cv2.resize(img, (s1,s2))
        except:
            raise Exception("Provide size in a format (s1,s2)")
    return img
```

## Watershed

```python
img = cv2.imread('./DATA/pennies.jpg')

# make blure
img = cv2.medianBlur(img, 35)

# make gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# make threshold
thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)[1]

# denoise by morphology
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)

# selecting regions that we are sure about (centers of coins)
sure_bg = cv2.dilate(opening,kernel,iterations=3)

dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)[1]

# regions that yet to be confirmed
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)

# make markers for watershed to start with
markers = cv2.connectedComponents(sure_fg)[1]

# Add one to all labels so that sure background is not 0, but 1
markers = markers+1

# Now, mark the region of unknown with zero
markers[unknown==255] = 0
markers = cv2.watershed(img,markers)

contours, hierarchy = cv2.findContours(markers.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

for i in range(len(contours)):
    
    if hierarchy[0][i][3] == -1:
        
        cv2.drawContours(img, contours, i, (255, 0, 0), 10)

plt.imshow(img)
```

# RemoveLable

```python
import tensorflow as ts
from tensorflow import keras
import glob
```

```python
img_s = 64
X,Y = [],[]
img_back = read_img('./DATA/watermark_no_copy.png', size=(img_s,img_s))
file_names = glob.glob("./DATA/pokemon/*.jpg")
for file in file_names:
    file = file.split("/")[-1].split(".")[0]
    try:
        file = int(file)
        img = read_img(f"./DATA/pokemon/{file}.jpg", size=(img_s,img_s))
        blended = cv2.addWeighted(src1=img, alpha=0.9, src2=img_back, beta=0.1, gamma=0)
        img_gray = cv2.cvtColor(blended, cv2.COLOR_BGR2GRAY)
        X.append(img_gray/255.)
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        Y.append(img_gray/255.)
    except:
        pass

X = np.array(X).reshape(-1,img_s,img_s,1)
Y = np.array(Y).reshape(-1,img_s,img_s,1)

X_train, X_test, y_train, y_test = train_test_split(X,Y)
```

```python
input_shape = (img_s, img_s, 1)
latent_dim = 128

# Encoder
inputs = keras.layers.Input(shape=input_shape)
x = keras.layers.Conv2D(32, kernel_size=3, strides=2, activation='relu', padding='same')(inputs)
x = keras.layers.Conv2D(64, kernel_size=3, strides=2, activation='relu', padding='same')(x)
x = keras.layers.Flatten()(x)
latent_repr = keras.layers.Dense(latent_dim)(x)

# Decoder
x = keras.layers.Dense(16*16*64)(latent_repr)
x = keras.layers.Reshape((16, 16, 64))(x)
x = keras.layers.Conv2DTranspose(32, kernel_size=3, strides=2, activation='relu', padding='same')(x)
decoded = keras.layers.Conv2DTranspose(1, kernel_size=3, strides=2, activation='sigmoid', padding='same')(x)

# Autoencoder model
autoencoder = keras.models.Model(inputs, decoded)

# compile
autoencoder.compile(optimizer=keras.optimizers.Adam(lr=0.0002), loss='binary_crossentropy')
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
epochs = 150
batch_size = 32

history = autoencoder.fit(X_train, y_train, validation_split=0.2,
                          epochs=epochs, batch_size=batch_size, callbacks=[early_stopping])
```

```python
pred = autoencoder.predict(X_test[0].reshape(1,img_s,img_s,1))
plt.imshow(pred.reshape(img_s,img_s,1), cmap="gray")
```

## 128bland

```python
img_s, back_s = 128, 96
img_back = read_img('./DATA/watermark_no_copy.png', size=(back_s,back_s))
X,Y = [],[]
file_names = glob.glob("./DATA/pokemon/*.jpg")
for file in file_names:
    file = file.split("/")[-1].split(".")[0]
    try:
        file = int(file)
        img = read_img(f"./DATA/pokemon/{file}.jpg", size=(img_s,img_s))
        Y.append(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)/255.0)
        r1,r2 = np.random.randint(img_s-back_s), np.random.randint(img_s-back_s)
        roi = img[r1:r1+back_s, r2:r2+back_s]

        mask = cv2.cvtColor(img_back, cv2.COLOR_BGR2GRAY)
        mask_inv = cv2.bitwise_not(mask)

        white_background = np.full(img_back.shape, 255, dtype=np.uint8)
        bk = cv2.bitwise_or(white_background, white_background, mask=mask_inv)
        fg = cv2.bitwise_or(img_back, img_back, mask=mask_inv)

        final_roi = cv2.bitwise_or(roi,fg)

        img[r1:r1+back_s, r2:r2+back_s] = final_roi
        X.append(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)/255.0)
    except:
        pass

X = np.array(X).reshape(-1,img_s,img_s,1)
Y = np.array(Y).reshape(-1,img_s,img_s,1)

X_train, X_test, y_train, y_test = train_test_split(X,Y)
```

```python
plt.imshow(X_train[29], cmap="gray")
```

```python
input_shape = (img_s, img_s, 1)
latent_dim = 128

# Encoder
inputs = keras.layers.Input(shape=input_shape)
x = keras.layers.Conv2D(32, kernel_size=3, strides=2, activation='relu', padding='same')(inputs)
x = keras.layers.Conv2D(64, kernel_size=3, strides=2, activation='relu', padding='same')(x)
x = keras.layers.Conv2D(64, kernel_size=3, strides=2, activation='relu', padding='same')(x)
x = keras.layers.Flatten()(x)
latent_repr = keras.layers.Dense(latent_dim)(x)

# Decoder
x = keras.layers.Dense(16*16*64)(latent_repr)
x = keras.layers.Dropout(0.3)(x)
x = keras.layers.Dense(16*16*64)(latent_repr)
x = keras.layers.Reshape((16, 16, 64))(x)
x = keras.layers.Conv2DTranspose(64, kernel_size=3, strides=2, activation='relu', padding='same')(x)
decoded = keras.layers.Conv2DTranspose(1, kernel_size=3, strides=4, activation='sigmoid', padding='same')(x)

# Autoencoder model
autoencoder = keras.models.Model(inputs, decoded)

# compile
autoencoder.compile(optimizer=keras.optimizers.Adam(lr=0.0002), loss='binary_crossentropy')
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

#autoencoder.summary()
```

```python
epochs = 150
batch_size = 32

history = autoencoder.fit(X_train, y_train, validation_split=0.2,
                          epochs=epochs, batch_size=batch_size, callbacks=[early_stopping])
```

```python
img_n = 10
pred = autoencoder.predict(X_test[img_n].reshape(1,img_s,img_s,1))
plt.imshow(pred.reshape(img_s,img_s,1), cmap="gray")
```
